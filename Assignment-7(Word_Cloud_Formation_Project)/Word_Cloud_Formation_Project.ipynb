```python
#Importing the requisite libraries/packages:
import PyPDF2 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
```

```python
#Creating a file uploading widget using 'ipywidgets' library:
from ipywidgets import FileUpload
upload = FileUpload()
upload
```

```python
#Writing the data of the inputed file(present in bytes form) into a PDF format:
with open("input.pdf", "w+b") as i:
    i.write(upload.data[0])
```

```python
# Open the PDF file and save it as a variable
resume = open('input.pdf', 'rb')
```

```python
#Checking the final format of the saved 'resume' variable:
resume
```

```python
#Creating a PDF reader object:
pdfReader = PyPDF2.PdfFileReader(resume)
```

```python
#Inspecting the number of pages in the pdfReader object:
print(pdfReader.numPages)
```

```python
#Creating a page-object for extracting the text of the 'first page' of PDF file:
pageObj = pdfReader.getPage(0)
```

```python
#Printing the text present on the 'first page' of the PDF file:
resume_text = pageObj.extractText()
resume_text
```

```python
#Closing the opened PDF file:
resume.close()
```

```python
#Removing the Line change('\n') character from the text of PDF file:
resume_text=resume_text.replace('\n',"")
resume_text
```

```python
#Preparing a 800x800 'Square Word Cloud': 
resume_wordcloud =  WordCloud(width = 800, height = 800, background_color ='white', min_font_size = 10).generate(resume_text) 
resume_wordcloud
```

```python
#Pictoral representation of word cloud:                        
plt.figure(figsize=(8,8)) 
plt.imshow(resume_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad=0) 
plt.show()
```

<b>Note:</b> It is important to note here that currently out text contains numerous words which are used as connectives or starters in english language(such as <b>'and'</b>,<b>'or'</b> etc.) and are termed as <b>'STOPWORDS'</b>, which are not be included in the Word-Clouds.Hence we will next remove these 'Stopwords' from our text and then re-print our Word-Cloud.

```python
#Inspecting all the 'Stopwords' present in the 'STOPWORDS' library:
stopwords = set(STOPWORDS)
stopwords
```

```python
#Removing the 'Stopwords' from the text of our PDF file:
resume_wordcloud =  WordCloud(width=800,height=800,background_color='white',stopwords=stopwords,min_font_size = 10).generate(resume_text) 
resume_wordcloud
```

```python
#Pictorial representation of Word-Cloud without'Stopwords':                     
plt.figure(figsize=(8,8),facecolor=None) 
plt.imshow(resume_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```

<b>Note:</b> The close observation of the Word-Cloud shows that there are still some irrelevant words(such as<b> 'Pvt'</b>,<b>'Ltd'</b> etc.)which can be removed and the Word-Cloud can be represented better.

```python
#Listing the irrelevant words in our Word-Cloud:
irrelevant_words = ['user','Pride','etc','full','care','Have','nearby','area','sustains','Pvt','Ltd','.']
```

```python
#Removing the irrelevant words from the text of our PDF file:
for word in irrelevant_words:
    if word in resume_text:
        resume_text=resume_text.replace(word,"")
resume_text
```

```python
#Pictorial representation of the filtered Square Word-Cloud:
resume_wordcloud =  WordCloud(width=600,height=600,background_color ='white',stopwords=stopwords,min_font_size=10).generate(resume_text)                         
plt.figure(figsize=(8,8),facecolor=None) 
plt.imshow(resume_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad=0) 
plt.savefig('Word-Cloud.png')
plt.show()
```

<b>Note:</b> Now the Word-Cloud seems to be better and cleaner without the irrelevant words.
