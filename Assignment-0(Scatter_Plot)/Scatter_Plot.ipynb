```python
import pandas as pd
from matplotlib import pyplot as plt
df=pd.read_csv("advertisement.csv")
df.head()
```

```python
plt.title("Effect of Advertisement via TV on Sales",fontweight='bold',fontsize=14)
plt.plot(df.TV,df.Sales,'o')
plt.xlabel("Expenditure(in Lakhs) on Advertisement via TV",fontweight='bold',fontsize=12)
plt.ylabel("Sales(in Crores)",fontweight='bold',fontsize=12)
plt.legend(["Data Points"])
plt.savefig("SalesVsExpenditure(TV).png")
plt.show()
```

```python
plt.title("Effect of Advertisement via Radio on Sales",fontweight='bold',fontsize=14)
plt.plot(df.Radio,df.Sales,'o',color='r')
plt.xlabel("Expenditure(in Lakhs) on Advertisement via Radio",fontweight='bold',fontsize=12)
plt.ylabel("Sales(in Crores)",fontweight='bold',fontsize=12)
plt.legend(["Data Points"])
plt.savefig("SalesVsExpenditure(Radio).png")
plt.show()
```

```python
plt.title("Effect of Advertisement via Newspaper on Sales",fontweight='bold',fontsize=14)
plt.plot(df.Newspaper,df.Sales,'o',color='green')
plt.xlabel("Expenditure(in Lakhs) on Advertisement via Newspaper",fontweight='bold',fontsize=12)
plt.ylabel("Sales(in Crores)",fontweight='bold',fontsize=12)
plt.legend(["Data Points"])
plt.savefig("SalesVsExpenditure(Newspaper).png")
plt.show()
```

```python
plt.title("Effect of Advertisement via TV & Radio on Sales",fontweight='bold',fontsize=14)
plt.plot((df.Radio+df.TV),df.Sales,'o',color="purple")
plt.xlabel("Expenditure(in Lakhs) on Advertisement via TV & Radio",fontweight='bold',fontsize=12)
plt.ylabel("Sales(in Crores)",fontweight='bold',fontsize=12)
plt.legend(["Data Points"])
plt.savefig("SalesVsExpenditure(TV & Radio).png")
plt.show()
```

```python
plt.title("Effect of Advertisement via Radio & Newspaper on Sales",fontweight='bold',fontsize=14)
plt.plot((df.Radio+df.Newspaper),df.Sales,'o',color="orange")
plt.xlabel("Expenditure(in Lakhs) on Advertisement via Radio & Newspaper",fontweight='bold',fontsize=12)
plt.ylabel("Sales(in Crores)",fontweight='bold',fontsize=12)
plt.legend(["Data Points"])
plt.savefig("SalesVsExpenditure(Radio & Newspaper).png")
plt.show()
```

```python
plt.title("Effect of Advertisement via TV & Newspaper on Sales",fontweight='bold',fontsize=14)
plt.plot((df.Newspaper+df.TV),df.Sales,'o',color="skyblue")
plt.xlabel("Expenditure(in Lakhs) on Advertisement via TV & Newspaper",fontweight='bold',fontsize=12)
plt.ylabel("Sales(in Crores)",fontweight='bold',fontsize=12)
plt.legend(["Data Points"])
plt.savefig("SalesVsExpenditure(TV & Newspaper).png")
plt.show()
```
